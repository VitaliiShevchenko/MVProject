# INSTALL DEPENDENCIES

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
colorImgBgr = cv2.imread(r'../../Downloads/download.png')
grayImg = cv2.imread(r'../../Downloads/puppy.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(grayImg, cmap='gray')
```

# ROTATION IMPLEMENTATION USING OPENCV

```python
image_center = tuple(np.array(grayImg.shape[::-1])/2)
degree = 45
scale = .7
rot_mat = cv2.getRotationMatrix2D(image_center, degree, scale)
img = cv2.warpAffine(grayImg, rot_mat, grayImg.shape[::-1])
plt.imshow(img, cmap='gray')
```
