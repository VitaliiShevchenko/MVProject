```python
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

import cv2
```

```python
im1 = cv2.imread(r'../../Downloads/1.jpg')
im2 = cv2.imread(r'../../Downloads/2.jpg')
im3 = cv2.imread(r'../../Downloads/3.jpg')
```

```python
plt.subplot(131)
plt.imshow(im1[:,:,::-1])
plt.subplot(132)
plt.imshow(im2[:,:,::-1])
plt.subplot(133)
plt.imshow(im3[:,:,::-1])
```

```python
imgs = []
imgs.append(im1)
imgs.append(im2)
imgs.append(im3)
M = cv2.Stitcher.create(cv2.Stitcher_PANORAMA)
status, pano = M.stitch(imgs)
plt.imshow(pano[:,:,::-1])
```
