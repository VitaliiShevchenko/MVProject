# INSTALL DEPENDENCIES

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r'../../Downloads/download.png')
img = img[:,:,::-1]
plt.imshow(img)
```

```python
green = np.uint8([[[0, 255, 0]]])
hsvGreen = cv2.cvtColor(green, cv2.COLOR_BGR2HSV)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
print("BGR of Green:", green)
print("HSV of Green:", hsvGreen)
```

```python
lr = np.array([50, 100, 100])
ur = np.array([70, 255, 255])
```

```python
mask_green = cv2.inRange(hsv, lr, ur)
```

```python
res = cv2.bitwise_and(img, img, mask = mask_green)
```
