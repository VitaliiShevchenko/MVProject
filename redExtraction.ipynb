```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread(r'../../Downloads/download.png')
img = img[:,:,::-1]
plt.imshow(img)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
lr = np.array([0, 120, 70])
ur = np.array([5, 255, 255])
```

```python
mask1 = cv2.inRange(hsv, lr, ur)
```

```python
mask1.shape
```

```python
lr = np.array([170, 120, 70])
ur = np.array([175, 255, 255])
```

```python
mask2 = cv2.inRange(hsv, lr, ur)
```

```python
np.unique(mask1)
```

```python
mask = mask1 | mask2
```

```python
res = cv2.bitwise_and(img, img, mask = mask)
```
