# OPENCV reads as BGR

```python
cimCv2 = cv2.imread(r'Downloads/valeriia-zm-rjkmn.jpg')
cimCv2 = cimCv2[:,:,::-1]

Rcv2 = cimCv2[:, :, 0]
Gcv2 = cimCv2[:, :, 1]
Bcv2 = cimCv2[:, :, 2]
```

```python
plt.figure(1)

plt.subplot(231)
plt.imshow(cimCv2)
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(cimCv2)
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(cimCv2)
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(Rcv2, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('RED')

plt.subplot(235)
plt.imshow(Gcv2, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('GREEN')

plt.subplot(236)
plt.imshow(Bcv2, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('BLUE')

plt.show()
```
